## univariate feature selection

```python
from pandas import read_csv
from numpy import set_printoptions
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
filename = startups
names = ['r&d_spend', 'Administration', 'marketing_spend', 'State', 'Profit']
```

```python
dataframe = (filename,names)
array= filename.values
```

```python
x = array[:,0:4]
y = array[:,4]
```

```python
#feature selection
test = SelectKBest(score_func=chi2, k= 4)
fit= test.fit(x,y)
```

```python
set_printoptions(precision = 3)
print(fit.scores_)
features = fit.transform(x)
```

```python
#according to this test marketing_spend is the best feature and in second no. 
#there is r&d and 3rd administration we can leave out state as it is ranked last
```

## feature selection using f_classif

```python
from sklearn.feature_selection import f_classif
test = SelectKBest(score_func = f_classif, k = 4)
fit = test.fit(x,y)
```

```python
set_printoptions(precision=3)
print(fit.scores_)
features = fit.transform(x)
```

## recursive feature elimination
